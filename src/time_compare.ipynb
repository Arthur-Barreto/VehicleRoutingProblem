```python
import matplotlib.pyplot as plt
import numpy as np
import os
import time
```

## Run global search for each file on `/input_files` and plot the time taken to each one

```python
input_file_folder = os.path.join(os.getcwd(), 'input_files')
```

```python
global_results = {}

for file in os.listdir(input_file_folder):
    # file format: grafoINT.txt
    # take the number from the file name
    graph_number = int(file[5:-4])
    start = time.time()
    # call the exe global_search with the input file
    os.system(f'./global_search {os.path.join(input_file_folder, file)}')
    end = time.time()
    global_results[graph_number] = end - start
```

```python
plt.plot(list(global_results.keys()), list(global_results.values()), 'ro')
plt.xlabel('Graph number')
plt.ylabel('Execution time')
plt.title('Global search execution time')
plt.show()
```

## Run parallel_seach for each file on `/input_files` and plot the time taken to each one

```python
parallel_search = {}

for file in os.listdir(input_file_folder):
    # file format: grafoINT.txt
    # take the number from the file name
    graph_number = int(file[5:-4])
    start = time.time()
    # call the exe parallel_search with the input file
    os.system(f'./parallel_search {os.path.join(input_file_folder, file)}')
    end = time.time()
    parallel_search[graph_number] = end - start
```

```python
plt.plot(list(parallel_search.keys()), list(parallel_search.values()), 'ro')
plt.xlabel('Number of Nodes')
plt.ylabel('Execution time (s)')
plt.title('Parallel search execution time')
plt.show()
```

### Compare the time taken for each file on both methods

```python
plt.plot(list(global_results.keys()), list(global_results.values()), 'ro', label='Global search')
plt.plot(list(parallel_search.keys()), list(parallel_search.values()), 'bo', label='Parallel search')
plt.xlabel('Number of Nodes')
plt.ylabel('Execution time (s)')
plt.title('Execution time comparison')
plt.legend()
```
