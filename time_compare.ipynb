```python
import matplotlib.pyplot as plt
import os
import time
import pandas as pd
import numpy as np
```

## Run global search for each file on `/input_files` and plot the time taken to each one

```python
input_file_folder = os.path.join(os.getcwd(), "input_files")
```

```python
capacity = 15
```

```python
global_results = {}

for file in os.listdir(input_file_folder):
    # file format: grafoINT.txt
    # take the number from the file name
    graph_number = int(file[5:-4])
    start = time.time()
    # call the exe global_search with the input file
    os.system(f"./global_search {os.path.join(input_file_folder, file)} {capacity}")
    end = time.time()
    global_results[graph_number] = end - start
```

```python
plt.plot(list(global_results.keys()), list(global_results.values()), "ro")
plt.xlabel("Graph number")
plt.ylabel("Execution time (s)")
plt.title("Global search execution time")
plt.show()
```

## Read the log file from `MPI`, the cluster solution and plot the time taken to each one

The file `mpi_time.csv` contains the time taken to run the `MPI` solution on the cluster. To more information about the time for each file, go to the folder `cluster_log` and read the log file. 

```python
data = {
    "graph_order": [4, 5, 6, 7, 8, 9, 10, 11, 12],
    "time_taken": [1.178, 1.179, 1.166, 1.167, 1.186, 1.238, 1.804, 7.583, 76.535]
}
df = pd.DataFrame(data)
```

```python
plt.plot(df["graph_order"], df["time_taken"], "ko")
plt.xlabel("Number of Nodes")
plt.ylabel("Execution time (s)")
plt.title("MPI search execution time")
plt.show()
```

## Compare the time taken for all the methods

```python
plt.figure(figsize=(10, 7))

plt.plot(
    list(global_results.keys()),
    np.log10(list(global_results.values())),
    "rx",
    label="Global search",
)
plt.plot(
    list(heuristic_results.keys()),
    np.log10(list(heuristic_results.values())),
    "gx",
    label="Heuristic search",
)
plt.plot(
    list(parallel_search.keys()),
    np.log10(list(parallel_search.values())),
    "bx",
    label="Parallel search",
)
plt.plot(df["graph_order"], np.log10(df["time_taken"]), "kx", label="MPI search")
plt.xlabel("Number of Nodes")
plt.ylabel("Execution time [log10(s)]")
plt.title("Execution time comparison")
plt.legend()
plt.plot()
```
